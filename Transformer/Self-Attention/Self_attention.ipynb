```python
import numpy as np
```

```python
# 3 tokens, embedding size 4
X = np.array([
    [1, 0, 1, 0],   # "I"
    [0, 2, 0, 2],   # "love"
    [1, 1, 1, 1]    # "AI"
])

print("Input Shape:", X.shape)
print(X)
```

```python
np.random.seed(0)

# 3 words, each has 4 numbers (embedding size = 4)
X = np.random.rand(3, 4)

print("Input Shape:", X.shape)
print(X)

## each row is a token vector
```

```python
# Dot product similarity
scores = X @ X.T
print("Raw Attention Scores:\n", scores)
```

```python
print(scores.max())
```

```python
np.random.seed(0)  # for reproducibility

d_model = 4  # embedding size

# Random weight matrices (learnable in real models)
Wq = np.random.rand(d_model, d_model)
Wk = np.random.rand(d_model, d_model)
Wv = np.random.rand(d_model, d_model)

# Compute Q, K, V
Q = X @ Wq
K = X @ Wk
V = X @ Wv

print("Q shape:", Q.shape)
print("K shape:", K.shape)
print("V shape:", V.shape)
```

```python
scores = Q @ K.T  ## Each word now has score with every word.
print("Raw Attention Scores:\n", scores)
```

```python
dk = d_model
scaled_scores = scores / np.sqrt(dk)
print('Scaled Scores:\n',scaled_scores)
```

```python
def softmax(x):
    exp_x = np.exp(x)
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)

attention_weights = softmax(scores)

print(attention_weights)
```

```python
output = attention_weights @ V
print('Self-Attention Output:\n', output)
```
